```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def denom(x):
    return 3*x**2 + 2
    
def y(x):
    return (x*(x**2 + 6)) / denom(x)

def y_minus_x(x):
    return (2*x*(2 - x**2)) / denom(x)

def y_squared_minus_2(x):
    d = denom(x)
    return (x**2 - 2)**3 / d**2

def y_squared(x):
    return y(x)**2
```

```python
# Generate x values.
x_values = np.linspace(-5, 5, 100)

# Generate x^2 values.
x_squared = np.square(x_values)

# Generate y values for y^2.
y_squared_values = y_squared(x_values)

# Generate y values for y^2 - 2.
y_squared_minus_2_values = y_squared_minus_2(x_values)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(9, 4))

# Plot y-x.
y_minus_x_values = y_minus_x(x_values)
axs[1].plot(x_values, y_minus_x_values, label='$y-x$')

# Plot x^2.
x_squared = np.square(x_values)
axs[0].plot(x_values, x_squared, label='$x^2$')

# Plot y^2.
axs[0].plot(x_values, y_squared_values, label='$y^2$')

# Plot y^2 - 2.
# Plot y^2 where y^2 - 2 is less than 0.
axs[1].plot(x_values[y_squared_minus_2_values < 0], y_squared_minus_2_values[y_squared_minus_2_values < 0], 'r-', label='$y^2$ (y^2 - 2 < 0)')

# Plot y^2 where y^2 - 2 is greater than or equal to 0.
axs[1].plot(x_values[y_squared_minus_2_values >= 0], y_squared_minus_2_values[y_squared_minus_2_values >= 0], 'r.', label='$y^2$ (y^2 - 2 >= 0)')


axs[0].axvline(x=np.sqrt(2), color='black', linestyle='dashed', linewidth=0.5)
axs[0].axhline(y=np.sqrt(2), color='black', linestyle='dashed', linewidth=0.5)

for ax in axs:
    ax.axhline(0, color='black', linewidth=1.5)
    ax.axvline(0, color='black', linewidth=1.5)

    # Add labels and title.
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()
    ax.grid(True)

axs[0].set_xlim((-2, 3))
axs[0].set_ylim((-1, 4))

axs[1].set_xlim((-3, 3))
#axs[1].set_ylim((-1, 4))

plt.tight_layout()
plt.show()
```

```python
from sympy import symbols, diff, solve, solve_poly_system, solve_univariate_inequality, Eq
```

```python
# Define the variable.
x = symbols('x')

y = (x*(x**2 + 6)) / (3*x**2 + 2)
y_squared = y**2

y_squared_minus_2 = (x**2 - 2)**3 / (3*x**2 + 2)**2

# Compute the derivative of y^2 with respect to x.
dy_squared_dx = diff(y_squared, x)

# Compute the second derivative of y^2 with respect to x.
d2y_squared_dx2 = diff(dy_squared_dx, x)

# Compute the derivative of y^2 - 2 with respect to x.
dy_squared_minus_2_dx = diff(y_squared_minus_2, x)

# Compute the second derivative of y^2 - 2 with respect to x.
d2y_squared_minus_2_d2x = diff(dy_squared_minus_2_dx, x)


# Compute the critical points of y^2.
critical_points_y_squared = solve(dy_squared_dx, x)

# Compute the inflection points of y^2.
inflection_points_y_squared = solve(d2y_squared_dx2, x)

# Compute the critical points of y^2 - 2.
critical_points_y_squared_minus_2 = solve(dy_squared_minus_2_dx, x)

# Compute the inflection points of y^2.
inflection_points_y_squared_minus_2 = solve(d2y_squared_minus_2_d2x, x)
```
